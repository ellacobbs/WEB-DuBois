```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c3_data = pd.read_csv("./data/challenge-3.csv")
c3_data.head()
```

```python
c4_data = pd.read_csv("./data/challenge-4.csv")
c4_data.head()
```

```python
c7_data = pd.read_csv("./data/challenge-7.csv")
c7_data.head()
```

```python
# Challenge 3

# plot
c3_plot = sns.barplot(c3_data, x="Land", y="Date", orient="h", color="red")
c3_plot.set(xlabel="", ylabel="", xticklabels="", title="ACRES OF LAND OWNED BY BLACK PEOPLE IN GEORGIA")

# text
c3_plot.text(100000, 0.3, "338,769", fontsize=7)
c3_plot.text(450000, 24.2, "1,062,223", fontsize=7);
```

```python
# Challenge 4

values = c4_data["Percentage"].to_list()

fig, ax = plt.subplots(figsize=(4, 6))

# plot
ax.bar(0, values[2], width=0.5, color="gold", label="YELLOW") 
ax.bar(0, values[1], width=0.5, color="#38270C", bottom=values[2], label="BROWN") 
ax.bar(0, values[0], width=0.5, color="black", bottom=values[2] + values[1], label="BLACK") 

# text
ax.text(-0.03, 75, "44%", color="white")
ax.text(-0.03, 35, "40%", color="red")
ax.text(-0.03, 8, "16%")

ax.text(-.60, 90, "BLACK.", fontsize=15)
ax.text(-.55, 87, "I.E. FULL-BLOODED", fontsize=8)
ax.text(-.55, 84, "NEGROES.", fontsize=8)

ax.text(-.60, 50, "BROWN.", fontsize=15)
ax.text(-.55, 46, "I.E. PERSONS WITH", fontsize=8)
ax.text(-.55, 43, "SOME WHITE BLOOD", fontsize=8)
ax.text(-.55, 40, "OR DESCENDANTS", fontsize=8)
ax.text(-.55, 37, "OF LIGHT COLORED", fontsize=8)
ax.text(-.55, 34, "AFRICANS.", fontsize=8)

ax.text(-.60, 14, "YELLOW.", fontsize=15)
ax.text(-.55, 10, "I.E. PERSONS WITH", fontsize=8)
ax.text(-.55, 7, "MORE WHITE THAN", fontsize=8)
ax.text(-.55, 4, "NEGROE BLOOD.", fontsize=8)

# clear axes
ax.set_xticks([])
ax.set_yticks([])

# titles
ax.set_title("BASED ON A STUDY OF 40,000 INDIVIDUALS OF NEGRO DESCENT.", fontsize=10)
fig.suptitle("RACE AMALGATION IN GEORGIA", y=0.94);
```

```python
# Challenge 7

fig, ax = plt.subplots(figsize=(6, 4))

# set colors
colors = ["red" if y == "Negroes, U.S.A." else "green" for y in c7_data["Country"]]

# plot
c7_plot = sns.barplot(c7_data, x="Illiteracy Rate", y="Country", orient="h", palette=colors, ax=ax)
c7_plot.set(xlabel="", ylabel="", xticklabels="")

# set titles
ax.set_title("Done by Atlanta University.", fontsize=9)
fig.text(-0.08, 1, "Illiteracy of the American Negroes compared with that of other nations.", fontsize=15)
fig.text(-0.02, .95, "Proportion d'illettres les Negres Americains comparee a celle des autres nations", fontsize=11);
```
